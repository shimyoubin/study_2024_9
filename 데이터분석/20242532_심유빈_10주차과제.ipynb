```python
import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)
male = []
female = []

name = input('찾고 싶은 지역의 이름을 알려주세요: ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            i = i.replace(',',"")
            male.append(-int(i))
        for i in row[106:207]:
            i=i.replace(',',"")
            female.append(int(i))
        break

f.close()
plt.barh(range(len(male)), male, color='b', label='male')
plt.barh(range(len(female)), female, color='r', label='female')

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

name = input('궁금한 지역의 이름을 입력하세요: ')

for row in data:
    if (name in row[0]):
        male =0
        female=0
        for i in row[3:104]:
            i = i.replace(',',"")
            male = male + int(i)
        for i in row[106:207]:
            i=i.replace(',',"")
            female = female + int(i)
        break

f.close()
labels  = ['male', 'female']
colors = ['skyblue', 'pink']

plt.pie([male, female], labels=labels, colors=colors, autopct='%.1f%%', startangle=90)
plt.show()
```
