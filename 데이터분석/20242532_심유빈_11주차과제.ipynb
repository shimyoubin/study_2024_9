```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

header = next(data)

male = []
female = []

name = input('궁금한 동네를 입력하세요: ')

for row in data :
    if name in row[0]:
        for i in range(3, 104):
            male.append(int(row[i].replace(',','')))
        for i in range(106, 207):
            female.append(int(row[i].replace(',','')))
        break
f.close()

plt.plot(male, label='male', color='b' )
plt.plot(female, label='female', color='r' )

plt.grid(True)
plt.legend()
plt.show()
        
```

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
next(data)

male = []
female = []
size = []

name = input('궁금한 동네를 입력해 주세요:')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            male.append(int(row[i].replace(',',"")))
            female.append(int(row[i + 103].replace(',',"")))
            size.append(math.sqrt(int(row[i].replace(',',"")) + int(row[i+103].replace(",",""))))
        break
f.close()

plt.figure(figsize=(10,5))
plt.scatter(male,female, c=range(len(male)), cmap='jet', alpha=0.7)
plt.colorbar()
plt.plot(range(max(male)), range(max(male)), 'g')
plt.grid(True)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
header = next (data)
next(data)

max_people = 0
max_station = ''
for row in data:
    for i in range(4, 52):
        row[i] = row[i].replace(',',"")

    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > max_people:
        max_people = sum(row[10: 15:2])
        max_station = row[3] +'(' + row[1] +')'

f.close()

print(max_station, ':', max_people, '명')
```
